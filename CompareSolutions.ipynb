```python
import json
with open("runbench5.3.json") as jfile:
    data = json.load(jfile)
data = data['data']
```

```python
from pprint import pprint
pts = [pt for pt in data if pt['word_range']==2**20]
pts_oct = (pt for pt in pts if pt['name']=='octavian')
pts_tido = (pt for pt in pts if pt['name']=='tido')
graph_oct = [(pt['docs'], min(pt['result'])/5) for pt in pts_oct if pt['words'] == 1024]
graph_tido = [(pt['docs'], min(pt['result'])/5) for pt in pts_tido if pt['words'] == 1024]
pprint(graph_oct)
pprint(graph_tido)
plot(*zip(*graph_oct), 'b', *zip(*graph_tido), 'r')
title('Runtime for set similarity with documents with length of 1024 words sampled from range(0, 2**20)')
xlabel('number of documents')
ylabel('time in seconds')
legend(('Octavian', 'Tido'))
```
